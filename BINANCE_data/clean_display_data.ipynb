```python
import pandas as pd
```

```python
df_doge=[]
df = pd.read_json('./DATA_new/April/json_data-28.json')
```

```python
df
```

```python
for i in range(len(df)):
    if df.iloc[i]['s'] == 'DOGEUSDT':
        df_doge.append(df.iloc[i]['k'])
```

```python
df_doge
```

```python
doge = pd.DataFrame(df_doge)
```

```python
doge_graph = doge.filter(['t','o'])
```

```python
doge_graph
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
from matplotlib.pyplot import figure
```

```python
for i in range(len(doge_graph)):
    doge_graph.iloc[i,0]=datetime.utcfromtimestamp((doge_graph.iloc[i,0])/1000).strftime('%H:%M')
    doge_graph.iloc[i,1]=float(doge_graph.iloc[i,1])

doge_graph
```

```python
figure(num=None,figsize=(18,10))
x = doge_graph['t'].tolist()
y = doge_graph['o'].tolist()
plt.xticks(ticks=range(len(y)), labels=x, rotation=80)
line, = plt.plot(x, y, lw=2)
```

```python
i=0
while i < 1440:
    doge_graph_portion = doge_graph.iloc[i:i+80]
    figure(num=None,figsize=(18,10))
    x = doge_graph_portion['t'].tolist()
    y = doge_graph_portion['o'].tolist()
    plt.xticks(ticks=range(len(y)), labels=x, rotation=80)
    display(plt.plot(x,y))
    i+=80
```
